```python
from gurobipy import *
m = Model()
v0 = m.addVar()
v1 = m.addVar()

m.update()
m.addConstr(v0-v1 <=4)
m.addConstr(v0+v1 <=4)
m.setObjective(-0.25*v0+v1,GRB.MAXIMIZE)
m.params.outputflag=0
m.optimize()
```

```python
import matplotlib.pyplot as p
p.plot([0,4],[0,4])
p.plot([4,0],[0,4])
p.plot([0,4],[1,2])
p.plot([v0.x],[v1.x],'ro')
p.show();
```
